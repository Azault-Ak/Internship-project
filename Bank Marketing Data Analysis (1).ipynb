```python
# Plotting the distribution of the target variabledf
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
# plot distribution of target variable y
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='y', hue='y', palette='Set2', legend=False)
plt.title('Distribution of Target Variable', fontsize=14)
plt.xlabel('Target (y)', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix for numerical features
```

```python
#.1 Select only numerical columns
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
#.2 Compute the correlation matrix
```

```python
correlation_matrix = numeric_df.corr()
```

```python
#.3 Plot the heat map
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={'shrink': 0.8})
plt.title('Correlation Matrix of Numerical Features', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
